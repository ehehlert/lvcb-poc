```python
import pandas as pd
```

```python
# Load the CSV files into DataFrames
data = pd.read_pickle('../models/cell-classification/training/balanced_set.pkl')
tags = pd.read_csv('../models/cell-classification/training/balanced_set_tags.csv')
```

```python
# Merge the 'filename' column from concatenated_tables onto T_clustered_tables based on 'table_id'
tagged_data = pd.merge(data, tags[['cell_id', 'cell_role']], on='cell_id', how='left')
```

```python
# Confirm it works
print(tagged_data.shape)
tagged_data.head()
```

```python
# Save the merged table to a new CSV file
tagged_data.to_pickle('tagged_cells_df.pkl')
```
